# Graph construction

```python
%matplotlib inline
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def graph_construction(n, p):
    ''' 
        n : number of nodes
        p : proportion of edges compare to complete graph
        
    '''
    G = nx.fast_gnp_random_graph(n,p)
    
    edge_weight_dict = {
        n:w for n, w in zip(G.edges(), map(lambda x: round(x,2),np.random.uniform(1.0,5.0, G.number_of_edges())))
    }
    
    for e in G.edges:
        G.edges[e]['weight'] = edge_weight_dict[e]
        
    return G
```

```python
G = graph_construction(50,0.2)
```
